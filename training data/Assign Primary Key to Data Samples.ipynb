```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("processed_audio_sample_scoring.xlsx")
```

```python
import uuid

def generate_unique_key():
    return str(uuid.uuid4())

# Example usage:
unique_key = generate_unique_key()
print(unique_key)
```

```python
df['primary key'] = df.apply(lambda x: generate_unique_key(), axis=1)
```
